Q1. Explain GET and POST methods.

#### Answer - 

GET method is used to send data through URL by the client and POST method is used to send data through the form or as a body by the client.

Q2. Why is request used in Flask?

#### Answer - 

In Flask request is used to ask a data from the user.

Q3. Why is redirect() used in Flask?

#### Answer - 

In Flask, the redirect() function is used to redirect the client's web browser to a different URL. 

Q4. What are templates in Flask? Why is the render_template() function used?

#### Answer - 

In Flask, templates are files that contain static content mixed with placeholders for dynamic data. The render_template() function is used to render the template to show the client side.
